```python
from myMLModule.models.cluster import MyKMeans
from sklearn.datasets import make_blobs
```

使用sklearn.datasets中的make_blobs生成数据

```python
data = make_blobs(n_samples=1000, n_features=2, centers=5, random_state=42)
X, y = data[0], data[-1]
print(X.shape)
print(y.shape)
```

使用matplotlib绘制

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(16, 10), dpi=60)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap="jet")
plt.show()
```

#### 使用kmeans模型聚类
##### 基本过程
1. 随机选取n个点作为初始中心点 (n为聚类数目)
2. 计算每个点到n个中心点的距离，取距离最近的点作为自己的聚类中心
3. 计算每个聚类中的平均值作为新的中心点
4. 若新的中心点与之前的中心点的偏移量小于阈值，则将取其为最终结果，否则继续迭代


```python
model = MyKMeans(n_clusters=5, n_epoches=1000)
model.fit(X)
pred_labels = model.labels

plt.figure(figsize=(16, 10), dpi=100)
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap="jet")
plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=pred_labels, cmap="jet")
plt.show()
```
