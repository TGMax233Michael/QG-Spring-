#### 逻辑回归分类

```python
from models.linear_model import LogisticRegression
from sklearn.datasets import make_moons
from metrics.classification import Accuracy
from preprocessing.feature import polynomial_feature
from model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np
```

利用sklearn.datasets创建一个月牙分布的分类数据集

```python
data = make_moons(n_samples=1000, noise=0.1)

X, y = data[0], data[1]
print(X)
print(y)
```

以特征1为横轴，特征2为纵轴绘制散点图

```python
plt.figure(figsize=(16, 10), dpi=60)
plt.scatter(X[:, 0], X[:, -1], c=y, cmap="summer")
plt.show()
```

使用逻辑回归进行分类

```python
model = LogisticRegression()
model.fit(X, y)
y_pred = model.predict(X)
print(f"准确率: {Accuracy(y, y_pred)}")
plt.figure(figsize=(20, 10), dpi=60)
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, -1], c=y, cmap="summer")
plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, -1], c=y_pred, cmap="summer")
plt.show()
```

尝试通过提升阶数来拟合其非线性关系，但似乎提升阶数会使得预测准确率更低

```python
X_poly = polynomial_feature(X, degrees=3)
model = LogisticRegression()
model.fit(X_poly, y)
y_pred = model.predict(X_poly)
print(f"准确率: {Accuracy(y, y_pred)}")
plt.figure(figsize=(20, 10), dpi=60)
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, -1], c=y, cmap="summer")
plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, -1], c=y_pred, cmap="summer")
plt.show()
```
